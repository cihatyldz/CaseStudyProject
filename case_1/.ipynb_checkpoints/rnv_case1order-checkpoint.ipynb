### 1.The DS Manager of RNV would like to process this messy dataset called order_line_data. Could you please perform a data cleaning to the Order_line_data dataset? 

#### Data Cleaning

```python
# Reads the Excel sheet into a DataFrame, prints its shape, previews the first rows, and creates a summary table of each columnâ€™s 
# dtype and missing-value count/percentage sorted by missingness.
raw_df = pd.read_excel(EXCEL_PATH, sheet_name=SHEET_ORDER)
print(f"Data shape: {raw_df.shape}") 
display(raw_df.head())

info_df = pd.DataFrame({
    'dtype': raw_df.dtypes.astype(str),
    'missing': raw_df.isna().sum(),
    'missing_pct': (raw_df.isna().mean() * 100).round(2)
}).sort_values('missing', ascending=False)
display(info_df)
```

```python
# Recognizing data by looking at the numeric describe values in the data
raw_df.describe(include=[np.number]).T
```

```python
ESSENTIAL_COLS: List[str] = ["mkp_order_no", "sku"]
NUMERIC_CANDIDATES: List[str] = ["item_sale_price","item_tax","shipping_sale","total_per_sku","transport_cost_actual","gm_estimated"]
DATE_HINTS: List[str] = ["order_date","order_place_time","order_prep_time","order_delivery_time","order_total_time"]
CATEGORICAL_CANDIDATES: List[str] = ["line_type","mkp_name","supplier_name","order_status","line_status"]
```

```python
# auxiliary functions

## Prints a timestamped message for step-by-step, traceable logging
def log(msg: str) -> None:
    print(f"{msg}")


## Standardizes column names to snake_case and removes special characters for robust downstream code
def normalize_colname(col: str) -> str:
    col = col.strip().lower()
    col = re.sub(r"\s+", "_", col)
    col = re.sub(r"[^\w_]", "", col)
    return col


## Trims whitespace in text columns and converts "nan"/"None"/"" strings to real NaN for consistent missing-data handling
def trim_object_columns(df: pd.DataFrame) -> pd.DataFrame:
    obj_cols = df.select_dtypes(include=["object"]).columns
    for c in obj_cols:
        df[c] = df[c].astype(str).str.strip()
        df[c] = df[c].replace({"nan": np.nan, "None": np.nan, "": np.nan})
    return df


## Cleans messy numeric strings (thousands separators, currency symbols, spaces) and safely converts to numeric dtype
def smart_to_numeric(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip()
    s = s.replace({"": np.nan, "None": np.nan, "nan": np.nan})
    s = s.str.replace(r"[,\s]", "", regex=True)      # binlik, boÅŸluk
    s = s.str.replace(r"[^\d\.\-]", "", regex=True) # para/harf
    s = s.apply(lambda x: re.sub(r"(?<=\d)\.(?=.*\.)", "", x) if isinstance(x, str) else x)
    return pd.to_numeric(s, errors="coerce")


## Bulk-applies smart_to_numeric to the specified numeric candidate columns
def coerce_numeric_columns(df: pd.DataFrame, numeric_cols: List[str]) -> pd.DataFrame:
    for c in numeric_cols:
        if c in df.columns:
            df[c] = smart_to_numeric(df[c])
    return df


## Detects date/time-like columns via hints/name patterns and parses them to datetime with coercion
def infer_and_parse_dates(df: pd.DataFrame, hints: List[str]) -> pd.DataFrame:
    candidates = set([c for c in hints if c in df.columns])
    for c in df.columns:
        if "date" in c or c.endswith("_time"):
            candidates.add(c)
    for c in candidates:
        df[c] = pd.to_datetime(df[c], errors="coerce", infer_datetime_format=True)
    return df


## Verifies required columns exist; raises an error if missing to enforce the data contract
def check_essential_columns(df: pd.DataFrame, essential_cols: List[str]) -> None:
    missing = [c for c in essential_cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing columns: {missing}")

## Drops rows with missing values in required columns and returns the cleaned DataFrame plus the count removed
def drop_essential_nas(df: pd.DataFrame, essential_cols: List[str]) -> Tuple[pd.DataFrame, int]:
    before = len(df)
    df2 = df.dropna(subset=[c for c in essential_cols if c in df.columns])
    return df2, before - len(df2)
```

```python
# Cleaning Pipeline
df = raw_df.copy()
df = df.rename(columns={c: normalize_colname(c) for c in df.columns})
df = trim_object_columns(df)
df = infer_and_parse_dates(df, hints=DATE_HINTS)
df = coerce_numeric_columns(df, NUMERIC_CANDIDATES)

before = len(df)
df = df.drop_duplicates()
removed_dupes = before - len(df)

check_essential_columns(df, ESSENTIAL_COLS)
df, removed_essential_na = drop_essential_nas(df, ESSENTIAL_COLS)

log(f"Duplicate removed: {removed_dupes}")
log(f"Required field removed: {removed_essential_na}")
display(df.head())
```

```python
# Derived Metrics
if 'total_per_sku' in df.columns:
    tax = df['item_tax'] if 'item_tax' in df.columns else 0.0
    df['net_revenue'] = df['total_per_sku'] - tax
else:
    df['net_revenue'] = np.nan

if 'transport_cost_actual' in df.columns:
    df['gross_margin_est'] = df['net_revenue'] - df['transport_cost_actual']
else:
    df['gross_margin_est'] = np.nan

display(df[['total_per_sku','item_tax','net_revenue','transport_cost_actual','gross_margin_est']].head(10))
```

```python
# Data Quality Control
issues = {}

for col in ['item_sale_price','item_tax','shipping_sale','total_per_sku','net_revenue','gross_margin_est']:
    if col in df.columns:
        neg = df[df[col] < 0]
        if not neg.empty:
            issues[f"negative_{col}"] = len(neg)

if {'total_per_sku','item_tax'}.issubset(df.columns):
    bad_tax = df[df['item_tax'] > df['total_per_sku']]
    if not bad_tax.empty:
        issues['tax_gt_total'] = len(bad_tax)

if {'mkp_order_no','sku'}.issubset(df.columns):
    dup_key = df.duplicated(subset=['mkp_order_no','sku'], keep=False).sum()
    if dup_key > 0:
        issues['potential_line_duplicates_by_key'] = int(dup_key)

display(pd.Series(issues, name='count'))
```

```python
## keep the latest line per (mkp_order_no, sku) by order_date
if {'mkp_order_no','sku','order_date'}.issubset(df.columns):
    df = (df.sort_values('order_date')
            .groupby(['mkp_order_no','sku'], as_index=False)
            .tail(1)
            .reset_index(drop=True))
```

```python
## Recalculate derived metrics saving
if 'total_per_sku' in df.columns:
    tax = df['item_tax'].fillna(0) if 'item_tax' in df.columns else 0
    df['net_revenue'] = df['total_per_sku'] - tax
else:
    df['net_revenue'] = np.nan

if 'transport_cost_actual' in df.columns:
    transport = df['transport_cost_actual'].fillna(0)
    df['gross_margin_est'] = df['net_revenue'] - transport
else:
    df['gross_margin_est'] = np.nan

df.to_csv('orders_clean_final.csv', index=False)
```

### 2.Could you carry out an EDA analysis on the dataset? We want to understand it.

#### EDA

```python
## Time Series Visualization
SAVE_MONTHLY_PNG = "orders_monthly_revenue.png"

kpis = {}
kpis['rows'] = len(df)
kpis['date_min'] = str(df['order_date'].min()) if 'order_date' in df.columns else None
kpis['date_max'] = str(df['order_date'].max()) if 'order_date' in df.columns else None
kpis['unique_orders'] = int(df['mkp_order_no'].nunique()) if 'mkp_order_no' in df.columns else None
kpis['unique_skus'] = int(df['sku'].nunique()) if 'sku' in df.columns else None
kpis['total_revenue'] = float(df['total_per_sku'].sum()) if 'total_per_sku' in df.columns else None
kpis['total_net_revenue'] = float(df['net_revenue'].sum()) if 'net_revenue' in df.columns else None
kpis['total_gm_est'] = float(df['gross_margin_est'].sum()) if 'gross_margin_est' in df.columns else None

display(pd.Series(kpis, name='value'))

if 'order_date' in df.columns and 'total_per_sku' in df.columns:
    monthly = (df.assign(month=lambda d: d['order_date'].dt.to_period('M'))
               .groupby('month')['total_per_sku'].sum().reset_index())
    monthly['month'] = monthly['month'].astype(str)

    plt.figure()
    plt.plot(monthly['month'], monthly['total_per_sku'])
    plt.title('Monthly Revenue (Total per SKU)')
    plt.xlabel('Month')
    plt.ylabel('Revenue')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(str(SAVE_MONTHLY_PNG))
    plt.show()
```

```python
# Top SKUs by revenue
top_skus = (df.groupby('sku', dropna=False)['total_per_sku']
              .sum().sort_values(ascending=False).head(10))
display(top_skus)

# Revenue by marketplace / supplier (if present)
if 'mkp_name' in df.columns:
    rev_by_mkp = df.groupby('mkp_name')['total_per_sku'].sum().sort_values(ascending=False)
    display(rev_by_mkp)
if 'supplier_name' in df.columns:
    rev_by_sup = df.groupby('supplier_name')['total_per_sku'].sum().sort_values(ascending=False)
    display(rev_by_sup)

# Order status distribution
if 'order_status' in df.columns:
    display(df['order_status'].value_counts(dropna=False))

# Daily trend (helps explain the partial October)
if 'order_date' in df.columns:
    daily = df.set_index('order_date').resample('D')['total_per_sku'].sum()
    display(daily)

# Average order value (AOV)
aov = (df.groupby('mkp_order_no')['total_per_sku'].sum()).mean()
print("AOV:", aov)

# Missingness profile
na = df.isna().mean().sort_values(ascending=False)
display(na)

# Outliers in revenue per line (IQR)
q1,q3 = df['total_per_sku'].quantile([0.25,0.75]); iqr = q3-q1
outliers = df[(df['total_per_sku'] < q1-1.5*iqr) | (df['total_per_sku'] > q3+1.5*iqr)]
print("Outlier lines:", len(outliers))

# Lead time (if timestamps exist)
if {'order_place_time','order_delivery_time'}.issubset(df.columns):
    df['order_total_hours'] = (df['order_delivery_time'] - df['order_place_time']).dt.total_seconds()/3600
    print("Lead time (h) â€” median:", df['order_total_hours'].median())
```
